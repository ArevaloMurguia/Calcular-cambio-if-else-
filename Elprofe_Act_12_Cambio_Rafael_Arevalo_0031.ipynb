## Las Variables
1. CP --> Costo del producto
2. CD --> Cantidad de dinero
3. Cambio

## Formulas
$Cambio = CD - CP$

## Diagrama de flujo LucidChat
<img src="https://raw.githubusercontent.com/ArevaloMurguia/Calcular-cambio-if-else-/refs/heads/main/DF%20if%2C%20else%20Cambio%202.png" width="350" height="500" alt="Texto descriptivo">


```python
# Para cuando es verdadero (Si)
# Calculando el cambio usando if else
# Datos de entrada CD , CP
# Mostrar cambio si es falso
# si es verdadero mostrar No alcanza
CP = float(input("Ingrese el costo del producto: "))
CD = float(input("Ingrese la cantidad de dinero: "))
if CP > CD:
    print("No alcanza")
else:
  Cambio = CD - CP
  print("El cambio es: ", Cambio)
print("Fin del programa")
print("Programa realizado por Arevalo Rafael NL = 3")
```

```python
# Para cuando es falso (No)
# Calculando el cambio usando if else
# Datos de entrada CD , CP
# Mostrar cambio si es falso
# si es verdadero mostrar No alcanza
CP = float(input("Ingrese el costo del producto: "))
CD = float(input("Ingrese la cantidad de dinero: "))
if CP > CD:
    print("No alcanza")
else:
  Cambio = CD - CP
  print("El cambio es: ", Cambio)
print("Fin del programa")
print("Programa realizado por Arevalo Rafael NL = 3")
```
